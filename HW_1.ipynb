```python
def problem_5(name_stock,cash = 100000, num_shares = 0, date_T = '2021-05-31',t = '2020-06-01'):
  start_date = '2019-06-01'
  end_date = '2021-05-31'
  data_stock = data.DataReader(name_stock, 'yahoo', start_date, end_date).reset_index(drop=False)
  date_T = datetime.datetime.strptime(date_T,"%Y-%m-%d")
  t = datetime.datetime.strptime(t,"%Y-%m-%d")
  data_stock['15'] = data_stock['Adj Close'].rolling(15).mean()
  data_stock['50'] = data_stock['Adj Close'].rolling(50).mean()
  data_stock['B_t'] = data_stock['15'] - data_stock['50']
  for i in range(1,len(data_stock.index)):
    while (data_stock.iloc[i]['Date'] <= date_T and data_stock.iloc[i-1]['Date'] >= t) and (cash > 0 or num_shares >0):
        if data_stock.iloc[i]['B_t'] > 0 and data_stock.iloc[i-1]['B_t'] < 0:
            if cash > 0:
                num_shares = cash/data_stock.iloc[i]['Adj Close']
                cash = 0
        elif data_stock.iloc[i]['B_t'] < 0 and data_stock.iloc[i-1]['B_t'] > 0:
            if num_shares > 0:
                cash = num_shares*data_stock.iloc[i]['Adj Close']
                num_shares = 0
        break
        
                
  cash = num_shares * data_stock.iloc[-1]['Adj Close']
  R = (cash - 100000)/100000
  print("Return %s = %s" %(name_stock,R))
```

#Problem 6

```python
start_date = '2018-06-01'
end_date = '2021-05-31'
FB = data.DataReader('FB', 'yahoo', '2018-05-01', end_date)
FB.sort_index(ascending=False, inplace=True)
FB['returns'] = (np.log(FB.Close / FB.Close.shift(-1)))
FB_daily_std = np.std(FB[(FB.index >= start_date) & (FB.index <= end_date)].returns)
FB_std = FB_daily_std * 252 ** 0.5
print('Daily volatility of FB: %s ,\nThe annualized volatility of FB: %s\n' %(FB_daily_std,FB_std))

GOOG = data.DataReader('GOOG', 'yahoo', '2018-05-01', end_date)
GOOG.sort_index(ascending=False, inplace=True)
GOOG['returns'] = (np.log(GOOG.Close / GOOG.Close.shift(-1)))
GOOG_daily_std = np.std(GOOG[(GOOG.index >= start_date) & (GOOG.index <= end_date)].returns)
GOOG_std = GOOG_daily_std * 252 ** 0.5
print('Daily volatility of GOOG: %s ,\nThe annualized volatility of GOOG: %s' %(GOOG_daily_std,GOOG_std))
```

```python
from numpy import cov
covariance = cov(FB[(FB.index >= start_date) & (FB.index <= end_date)].returns, GOOG[(GOOG.index >= start_date) & (GOOG.index <= end_date)].returns)
correlation_coefficient = covariance / (FB_daily_std * GOOG_daily_std)
print("The correlation coefficient ρ between the logarithmic returns of Facebook Inc. (FB) and Alphabet Inc. (GOOG) stocks: " ,correlation_coefficient[0,1])
```
